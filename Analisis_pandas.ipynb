# 3. Depuración de valores nulos

```python
# Visualización de datos nulos

df.isnull().sum() # Mostramos cuantos datos nulos hay en cada columna
```

Las columnas que presentan valores nulos son:
- Presión sistólica (14 valores nulos)
- Presión diastólica (16 valores nulos)
- Glucosa (14 valores nulos)
- Colesterol total (14 valores nulos)
- Colesterol LDL (13 valores nulos)
- Colesterol HDL (14 valores nulos)
- Respuesta (17 valores nulos)

```python
# Filtrado de datos nulos

df = df.dropna() # Eliminamos las filas con datos nulos en alguna de las columnas y definimos de nuevo la base de datos como df

df.isnull().sum() # Comprobamos que hemos eliminado los datos nulos
```

# 4. Creación de la variable índice de riesgo

```python
# Creación de la columna indice_riesgo

df["indice_riesgo"] = np.sqrt(df['glucosa']) + np.log(df['colesterol_total']) - (df['colesterol_hdl'] / 10)

# Comprobamos que se ha creado imprimiendo las 5 primeras columnas
df.head()
```

**Variable Indice_Riesgo:**
Riesgo de accidente cerebrovascular debido a
que se suman los factores de risgo glucosa y colesterol total y se resta
el colesterol HDL. Esto hace que si el índice es alto, se establezca un
mayor riesgo de ACV.

# 5. Cálculo de correlaciones entre dos columnas

```python
# Cálculo de correlaciones

# Comprobamos la correlación entre glucosa y colesterol LDL
pearson_gc = df["glucosa"].corr(df["colesterol_ldl"]) # Correlación lineal de Pearson. Para datos con distribución normal
spearman_gc = df["glucosa"].corr(df["colesterol_ldl"], method = "spearman") # Correlación monotonía. Para datos con tendencia no lineal y gran cantidad de datos. Se ve afectado por datos atípicos.
kendall_gc = df["glucosa"].corr(df["colesterol_ldl"], method = "kendall") # Correlación monotonía. Para datos con tendencia no lineal y poca cantidad de datos. Más robusto con datos atípicos.

# Comprobamos si hay correlación entre edad y respuesta
pearson_er = df["edad"].corr(df["respuesta"]) # Correlación lineal de Pearson. Para datos con distribución normal
spearman_er = df["edad"].corr(df["respuesta"], method = "spearman") # Correlación monotonía. Para datos con tendencia no lineal y gran cantidad de datos. Se ve afectado por datos atípicos.
kendall_er = df["edad"].corr(df["respuesta"], method = "kendall") # Correlación monotonía. Para datos con tendencia no lineal y poca cantidad de datos. Más robusto con datos atípicos.

# Informe de correlación glucos y colesterol LDL
print("================= Informe de correlación entre glusoca y colesterol LDL =================")
print(f'Pearson: {pearson_gc}')
print(f'Spearman: {spearman_gc}')
print(f'Kendall: {kendall_gc}')

# Informe de correlación edad y respuesta
print("\n================= Informe de correlación entre edad y respuesta =================")
print(f'Pearson: {pearson_er}')
print(f'Spearman: {spearman_er}')
print(f'Kendall: {kendall_er}')
```

**Conclusiones glucosa y colesterol LDL:**
No hay correlación entre la glucosa y el colesterol LDL, y de haberla es positiva y muy leve.
Debido a que hay una gran cantidad de datos, y que no presentan una distribución normal
porque presentan una desviación estándar grande. Nos quedamos con la correlación de Spearman.
Si existiera la correlación, querría decir que a mayor cantidad de glucosa en sangre, mayor
cantidad de colesterol LDL.

**Conclusiones edad y respuesta:**
No hay correlación entre la edad y el respuesta, y de haberla es positiva y muy leve.
Debido a que hay una gran cantidad de datos, y que no presentan una distribución normal
porque presentan una desviación estándar grande. Nos quedamos con la correlación de Spearman.
Si existiera la correlación, querría decir que a mayor edad, mayor respuesta.
