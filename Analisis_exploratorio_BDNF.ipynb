```python
# ===================== PACIENTES BDNF ======================
# Nombre: Anabel Gallego Pérez ==============================
# Descripción: Análisis exploratorio del BDNF en sangre =====
# Python 3.14.3 =============================================
# Jupyter lab ===============================================


# Librerías
import statistics as sts
import math

#Fucniones
#Función para cargar los datos en csv
def cargar_datos(archivo):
    """
    Función para abrir un archivo csv

    Parámetros:
    archivo (str)--> Nombre del archivo que queremos abrir en formato csv
    return--> valores_id (str): lista de cadena de texto con la identificación de los pacientes
          --> valores_bdnf (float): lista con los valores de BDNF en sangre (ng/mL).
    """
    with open (archivo + ".csv", "r") as archivo: # Abrimos el archivo de la base de datos
        lineas = archivo.readlines() # Leemos las líneas de la base de datos
    valores_id = [str(lineas[i].split(",")[0]) for i in range(1, len(lineas))] # Creamos la lista con ID
    valores_bdnf = [float(lineas[i].split(",")[1]) for i in range(1, len(lineas))] # Creamos la lista BDNF
    return valores_id, valores_bdnf # Devolvemos las dos listas de datos

# Función para cálculo estadístico descriptivo
def estadistica_descriptiva (datos):
    """
    Fucnión para el análisis estadístico descriptivo con librería statistics (sts)

    Parámetros:
    datos (list)--> Lista con datos numéricos (float) (int)
    return --> media (float): Media aritmética de la lista
           --> mediana (float): Mediana de la lista
           --> moda (float): Lista con la/s moda/s de la lista
           --> frecuencia_moda (int): Frecuencia en la que se repite la moda/s
           --> sd (float): Desviación estándar de la lista
           --> varianza (float): Varianza de la lista
           --> minimo (float): Valor mínimo de la lista
           --> maximo (float): Valor máximo de la ista
           --> q1 (float): Primer cuartil
           --> q2 (float): Segundo cuartil
           --> q3 (float): Tercer cuartil
    """
    media = sts.mean(datos) # Cálculo media
    mediana = sts.median(datos) # Cálculo mediana
    moda = sts.multimode(datos) # Cálculo lista con todas las modas
    frecuencia_moda = datos.count(moda[0]) # Cálculo de la frecuencia de la/s moda/s
    sd = sts.stdev(datos) # Cálculo desviación estándar
    varianza = sts.variance(datos) # Cálculo varianza
    minimo = min(datos) # Cálculo mínimo
    maximo = max(datos) # Cálculo máximo
    cuartiles = sts.quantiles(datos) # Cálculo de los cuartiles (Q1, Q2 y Q3)
    q1, q2, q3 = cuartiles # Definimos Q1, Q2 y Q3

    # Informe estadística descriptiva
    print(" =========== INFORME ESTADÍSTICO DESCRIPTIVO ===========")
    print(f'Media: {media}')
    print(f'Mediana: {mediana}')
    print(f'Modas: {moda}')
    print(f'Frecuencia de las modas: {frecuencia_moda}')
    print(f'Desviación estándar: {sd}')
    print(f'Varianza: {varianza}')
    print(f'Mínimo: {minimo}')
    print(f'Máximo: {maximo}')
    print(f'Primer Cuartil (Q1): {q1}')
    print(f'Segundo cuartil (Q2): {q2}')
    print(f'Tercer cuartil (Q3): {q3}')

# Definimos nuestro datos
# Creamos la lista con los ID
datos_id = cargar_datos("bdnf_pacientes")[0]
# Creamos la lista con los datos numéricos de BDNF en sangre (ng/mL)
datos_bdnf = cargar_datos("bdnf_pacientes")[1]


# Verificamos el funcionamiento de las listas de datos
print("Visualizamos 10 datos de cada lista")
# Verificamos la lista ID mostrando 10 datos
print(datos_id[:10])
#Verificamos la lista BDNF mostrando 10 datos
print(datos_bdnf[:10])


# Cálculo del total de lesionados en nuestra base de dato
total_id = len(datos_id) # Cálculo del total de datos de ID
total_bdnf = len(datos_bdnf) # Cálculo del toral de datos BDNF


# Comprobamos que tenemos los mismo datos en las dos listas
total_id == total_bdnf # True--> No hay pérdidas en la muestra
```

```python
# Calculamos estadística descriptiva
estadistica_descriptiva(datos_bdnf)
```

```python
# Análisis de vales atípicos

q1 = sts.quantiles(datos_bdnf)[0] # Primer cuartil
q3 = sts.quantiles(datos_bdnf)[2] # Segundo cuartil
iqr = q3 - q1 # Cálculo rango intercuartílico (IQR)
limite_inf = q1 - (1.5*iqr) # Límite inferior
limite_sup = q3 + (1.5*iqr) # Límite superior

# Identificación de los valores atípicos
valores_atipicos = [valor for valor in datos_bdnf if valor < limite_inf or valor > limite_sup]

# RETO: Identificamos el ID de los valores atípicos
# Identificamos la posición en la base de datos de BDNF
sitio_atipicos = [datos_bdnf.index(valor) for valor in valores_atipicos]
# Con la posición, identificamos el ID de los pacientes atípicos
id_atipicos = [datos_id[valor] for valor in sitio_atipicos]

# Informe de valores atípicos
print("============= INFORME VALORES ATÍPICOS =============")
print(f'IQR: {iqr}')
print(f'Límite inferior: {limite_inf}')
print(f'Límite superior: {limite_sup}')
print(f'Número de valores atípicos: {len(valores_atipicos)}')
print(f'Valores atípicos: {valores_atipicos}')
print(f'ID sujetos valores atípicos: {id_atipicos}')
```

```python
# Clasificación clínica de los valores de BDNF en plasmas:
# Bajo: < 20 ng/mL
# Normal: 20 – 40 ng/mL (incluye ambos extremos)
# Alto: > 40 ng/mL

cont_bajo = 0 # Total de pacientes con valores bajos
cont_normal = 0 # Total de pacientes con valores normales
cont_alto = 0 # Total de pacientes con valores altos

for valor in datos_bdnf:
    if valor < 20:
        cont_bajo = cont_bajo + 1
    elif valor > 40:
        cont_alto = cont_alto + 1
    else:
        cont_normal = cont_normal + 1

# Informe de clasificación clínica
print ("============== Clasificación clínica ==============")
print(f'Total pacientes: {len(datos_bdnf)}')
print(f'Pacientes con valores bajos: {cont_bajo}')
print(f'Pacientes con valores normales: {cont_normal}')
print(f'Pacientes con vales altos: {cont_alto}')
```

```python
# Prueba Estadística Inferencial
# Nivel del confianza del 95%

n = len(datos_bdnf) # Número total de sujetos
media = sts.mean(datos_bdnf) # Media muestral
sd = sts.stdev(datos_bdnf) # Desviación estándar
ee = sd/math.sqrt(n) # Error estándar (desviación estándar entre la raiz cuadrada del total de la muestra)
t_critico = 1.96 # El valor crítico para un intervalo de confianza al 95% es 1.96

# Intervalo de confianza
int_superior = media + (t_critico*ee)
int_inferior = media - (t_critico*ee)

# Prueba inferenciasl con intervalos de confianza al 95%
print("============= Prueba de intervalo de confianza al 95% =============")
print("Nivel de confianza: 95%")
print(f'Tamaño muestral: {n}')
print(f'Media: {sts.mean(datos_bdnf)}')
print(f'Error estandar: {ee}')
print(f'Intervalo de confianza: [{int_inferior},{int_superior}]')

if int_inferior >= 20 and int_superior <= 40:
    print("La media media poblacional de esta muestra está en rangos normales")
elif int_inferior < 20 and int_superior <= 40:
    print("La media poblacional de esta muestra está en rangos bajos-normales")
elif int_inferior < 20 and int_superior < 20:
    print("La media poblacional de esta muestra está en rangos bajos")
elif int_inferior >= 20 and int_superior > 40:
    print("La media poblacional de esta muestra está en rangos normales-altos")
else:
    print("La media poblacional de esta muestra está en rangos altos")
```
